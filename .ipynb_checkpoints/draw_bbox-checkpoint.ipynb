```python
from matplotlib import pyplot as plt
from PIL import Image
from glob import glob
import numpy as np
import ast
```

```python
def read_image(img_path):
    img = Image.open(img_path)
    return np.array(img)

def get_box(txt_path):
    data = open(txt_path).readlines()
    bbox = []
    for d in data:
        d = d.rstrip()
        d = np.array(ast.literal_eval(d))
        
        x1, x2 = max(d[:, 0]), min(d[:, 0])
        y1, y2 = max(d[:, 1]), min(d[:, 1])
        
        bbox.append([x1, y1, x2, y2])
    return bbox
```

```python
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image_path = sorted(glob("SegmentationImage/*.png"))
ped_box_path = sorted(glob("PedestrianBBox/*"))
veh_box_path = sorted(glob("VehicleBBox/*"))
```

```python
for i in range(len(ped_box_path)):
    print(i)
    plt.figure(figsize = (54/4, 96/4))
    img = read_image(image_path[i])

    bboxs = get_box(ped_box_path[i])
    for box in bboxs:
        plt.plot([box[0], box[0],box[2],box[2],box[0]], [box[1], box[3],box[3],box[1],box[1]])

    bboxs = get_box(veh_box_path[i])
    for box in bboxs:
        plt.plot([box[0], box[0],box[2],box[2],box[0]], [box[1], box[3],box[3],box[1],box[1]])
        
    plt.imshow(img)
    
    plt.show()
#     break
```
